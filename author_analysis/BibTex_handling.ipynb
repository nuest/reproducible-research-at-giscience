```python
from pybtex.database import parse_file
```

```python
# Scopus search query: 
# SOURCE-ID ( 21100244603 )  AND  CONFNAME ( agile ) 
bib_data = parse_file('AGILE_Conference_Scopus.bib')
```

```python
count = 0
for entry in bib_data.entries:
    count+=1
    print(count, entry)
```

```python
for author in bib_data.entries['Ludwig201165'].persons['author']:
    print(author)
```

```python
surnames_AGILE = []
count = 0
try:
    for entry in bib_data.entries:
        for author in bib_data.entries[entry].persons['author']:
            surname = str(author).split(',')[0]
            count+=1
            print(count, surname)
            if surname not in surnames_AGILE:
                surnames_AGILE.append(surname)
except:
    print("some error")
print(len(surnames_AGILE), surnames_AGILE)
```

```python
# Scopus search
# SOURCE-ID ( 25674 )  AND  CONFNAME ( giscience )  

bib_data = parse_file('GIScience_Conference_Scopus.bib')
surnames_GIScience = []
count = 0
try:
    for entry in bib_data.entries:
        for author in bib_data.entries[entry].persons['author']:
            surname = str(author).split(',')[0]
            count+=1
            print(count, surname)
            if surname not in surnames_GIScience:
                surnames_GIScience.append(surname)
except:
    print("some error")
print(len(surnames_GIScience), surnames_GIScience)
```

```python
surnames_AGILE_as_set = set(surnames_AGILE)
intersection = surnames_AGILE_as_set.intersection(surnames_GIScience)
intersection_as_list = list(intersection)
print(len(intersection_as_list), intersection_as_list)
```

Caveats:
- reliability of search in Scopus not checked: since one cannot search for conference well, combination of outlet and keyword used
- different authors with the same surname not differentiated
- only full papers; lots of the community interaction is with short papers or extended abstracts
